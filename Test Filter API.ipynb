```python
data = np.sin(np.linspace(0, 20*np.pi, num=300)) + np.cos(np.linspace(0, 10*np.pi, num=300)) + \
        + np.cos(np.linspace(0, 30*np.pi, num=300))
noise = np.random.randn(len(data))
rx_signal = data + noise
```

```python
plt.plot(rx_signal)
plt.hold(True)
plt.plot(data, c='r')
```

```python
data_str = str(rx_signal.tolist()).strip("[]")
ret = urllib.urlopen('http://{2}/filter/run?fid={0}&data={1}&padding=False'.format(fid, data_str, rooturl)).read()
```

```python
plt.figure(figsize=(12,4))
plt.plot(json.loads(ret)['data_in'].strip('[]').split(','))
plt.hold(True)
plt.plot(json.loads(ret)['data_out'])
plt.plot(data)
```

```python
# The filter introduces a delay between the input and the ouput. This is called "group delay".

x = np.fromstring(json.loads(ret)['data_in'].strip('[]'), sep=",")
y = np.array(json.loads(ret)['data_out'])    # TODO The return values are of different type. Not sure why.
```

```python
lags = plt.xcorr(x, y, maxlags=50)
plt.hold(True)
delay = lags[0][np.argmax(lags[1])]
print "Delay is {0} samples".format(delay)
```

The signal delay for a FIR filter is $\frac{(N_{taps} - 1)}{2}$

```python
plt.figure(figsize=(12,4))
plt.plot(range(len(data)) - delay, x)
plt.hold(True)
plt.plot(range(len(data)) - delay, data)
plt.plot(y)
```

```python
from scipy import signal
window=('kaiser', 8)
fir_coeff = signal.firwin(50, 0.05, window=window)
filtered_signal = signal.lfilter(fir_coeff, 1.0, data)
plt.figure()
plt.stem(fir_coeff)
plt.figure()
plt.plot(data)
plt.hold(True)
plt.plot(filtered_signal)
```

```python
# Modified from http://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.signal.square.html
from scipy import signal
import matplotlib.pyplot as plt
t = np.linspace(0, 3, 500, endpoint=False)
plt.plot(t, signal.square(2 * np.pi * 5 * t))
plt.ylim(-2, 2)

plt.figure()
sig = np.sin(2 * np.pi * t)
pwm = signal.square(2 * np.pi * 30 * t, duty=(sig + 1)/2)
plt.subplot(2, 1, 1)
plt.plot(t, sig)
plt.subplot(2, 1, 2)
plt.plot(t, pwm)
plt.ylim(-1.5, 1.5)
```
