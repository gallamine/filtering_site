```python
data = np.sin(np.linspace(0,40*np.pi,num=300)) + np.cos(np.linspace(0,105*np.pi,num=300)) + \
        + np.cos(np.linspace(0,30*np.pi,num=300))
```

```python
data_str = str(data.tolist())
ret = urllib.urlopen('http://localhost:5000/filter/run?fid={0}&data={1}'.format(fid,data_str)).read()
```

```python
plt.figure(figsize=(12,12))
plt.plot(json.loads(ret)['data_in'].strip('[]').split(','))
plt.hold()
plt.plot(json.loads(ret)['data_out'])
```

```python
from scipy import signal
window=('kaiser',8)
fir_coeff = signal.firwin(50, 0.05, window=window)
filtered_signal = signal.lfilter(fir_coeff,1.0, data)
plt.figure()
plt.stem(fir_coeff)
plt.figure()
plt.plot(data)
plt.hold(True)
plt.plot(filtered_signal)
```
